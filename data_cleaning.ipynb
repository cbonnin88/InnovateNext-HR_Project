# **Cleaning df_employees**

```python
# Cleaning df_employees
display(df_employees.head())
```

```python
# Getting todays date, will be usefu for the tenure calculation
today_date = datetime.now().date()
```

```python
df_employees_clean = df_employees.with_columns(
    # 1. Clean Department strings
    pl.col('Department')
    .str.to_lowercase() # ' sales' -> ' sales'
    .str.strip_chars()   # ' sales' -> 'sales'
    .str.to_titlecase()  #  'sales' -> 'Sales
    .replace('','Unknown') # '' -> 'Unknown'
    .replace('HR','Human Resources')
    .replace('hr','Human Resources')
    .replace('Hr','Human Resources')
    .replace('Nan','Unknown')
    .replace('NaN','Unknown')
    .fill_null('Unknown'), # null -> 'Unknown'

    # 2. Clean Gender
    pl.col('Gender')
    .replace('','Unknown')
    .fill_null('Unknown'),

    # 3. Clean Location
    pl.col('Location').str.strip_chars(),

    # 4. Convert strintg dates to Polars Date type
    # %d-%b-%Y format matches 'dd-Mon-YYYY' (e.g., 15-Mon-2020)
    pl.col('HireDate').str.to_date('%d-%b-%Y'),
    pl.col('TerminationDate').str.to_date('%d-%b-%Y', strict=False) # Add strict=False to handle 'NaN'
).unique(subset=['EmployeeID'], keep='first') # Remove duplicate EmployeeIDs

print('--- Cleaned Employees ---')
display(df_employees_clean.head())
```

```python
print(df_employees_clean.schema)
```

# **Cleaning df_salaries**

```python
df_salaries_clean = df_salaries.with_columns(
    # Use .str.replace_all() with a regex to remove '€' and ','
    pl.col('Salary')
    .str.replace_all(r'[€,]','') # Remove € and ,
    .cast(pl.Float64) # Cast to float (to keep the cents)
    .round(0) # Round to the nearest euro
    .cast(pl.Int64) # Cast to integer
)

print('--- Cleaned Salaries ---')
display(df_salaries.head())
```

```python
print(df_salaries.schema)
```

# **Cleaning df_engagement**

```python
df_engagement_clean = df_engagement.with_columns(
    # 1. Standardize EmployeeID
    pl.col('EmployeeID')
    .cast(pl.String) # Cast all to string first
    .map_elements(
        # Prepend 'E'
        lambda x: f'E{x}' if not x.startswith('E') else x,
        return_dtype = pl.String
    ),

    # 2. Clip outliers in scores
    pl.col('EngagementScore').clip(lower_bound=1,upper_bound=10),
    pl.col('ManagerRating').clip(lower_bound=1,upper_bound=10),

    # 3. Fix Date Type
    pl.col('SurveyDate').str.to_date('%Y-%m-%d')
)

print('--- Cleaned Engagement ---')
display(df_engagement_clean.head())
```

# **Cleaning df_performance**

```python
# Define the mapping for inconsistent ratings
rating_map = {
    'Low': '1',
    'Average':'3',
    'High':'5',
    'N/A': None # will become null
}

df_performance_clean=df_performance.with_columns(
    # 1. Map string ratings to numbers
    pl.col('PerformanceRating')
    .replace(rating_map)
    .cast(pl.Int64,strict=False), # strict=False turns errors (1,2,etc.) into null

    # 2. Fix ReviewYear type
    pl.col('ReviewYear').cast(pl.Int64, strict=False),

    # 3. Standardize Promotion
    pl.col('Promotion')
    .replace({'N/A':'No','':'No'})
    .fill_null('No')
)

print('--- Cleaned Performance ---')
display(df_performance_clean.head())
```

# **Aggregate, Join and Feature**

```python
# 1. Aggregate: Get *latest* performance review per employee
df_perf_latest = df_performance_clean.sort('ReviewYear',descending=True).group_by('EmployeeID').first()

# 2. Aggregate: Get *latest* engagement survey per employee
df_eng_latest = df_engagement_clean.sort('SurveyDate',descending=True).group_by('EmployeeID').first()
```

# **Joining everything to the main employee Table**

```python
df_final = (
    df_employees_clean
    # Join salaries (one-to-one)
    .join(df_salaries_clean, on='EmployeeID',how='left')
    # Join latest performance (now one-to-one)
    .join(
        df_perf_latest.select(['EmployeeID','PerformanceRating','Promotion']),
        on='EmployeeID',
        how='left'
    )
    # Join latest engagement (now one-to-one)
    .join(
        df_eng_latest.select(['EmployeeID','EngagementScore','ManagerRating']),
        on='EmployeeID',
        how='left'
    )
)
```

```python
# Final Feature Engineering
df_final = df_final.with_columns(
    # Create the target variable: Attrition (1 or 0)
    pl.when(pl.col('TerminationDate').is_null())
    .then(0)
    .otherwise(1)
    .alias('Attrition'),

    # Calculate Tenure
    pl.when(pl.col('TerminationDate').is_null())
    .then(
        # Current employee: today - HireDate
        (pl.lit(today_date) - pl.col('HireDate'))
    )
    .otherwise(
        # Former employee: TerminationDate - HireDate
        (pl.col('TerminationDate') - pl.col('HireDate'))
    )
    .dt.total_days()
    .cast(pl.Int64)
    .alias('TenureInDays')
)
```

```python
print('--- Final Combined & Cleaned DataFrame ---')
display(df_final.head())
print(f'\nFinal DataFrame shape: {df_final.shape}')
```
